```python
import cv2
from google.colab.patches import cv2_imshow
import mediapipe as mp
import time
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose

def distanceCalculate(p1, p2):
    """p1 and p2 in format (x1,y1) and (x2,y2) tuples"""
    dis = ((p2[0] - p1[0]) ** 2 + (p2[1] - p1[1]) ** 2) ** 0.5
    return dis




def pushUpCounter(path,res,name,maxDistance):
  pushUpStart = 0
  pushUpCount = 0
  cap = cv2.VideoCapture(path)
  result = cv2.VideoWriter(name + '.avi',cv2.VideoWriter_fourcc(*'MJPG'),25, res)
  with mp_pose.Pose(
      min_detection_confidence=0.5,
      min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
      success, image = cap.read()
      if not success:
        print("End of file.")
        break

      image.flags.writeable = False
      results = pose.process(image)
      image_height, image_width, _ = image.shape

      nosePoint = (int(results.pose_landmarks.landmark[0].x*image_width), int(results.pose_landmarks.landmark[0].y*image_height))
      leftWrist = (int(results.pose_landmarks.landmark[15].x*image_width), int(results.pose_landmarks.landmark[15].y*image_height))
      rightWrist = (int(results.pose_landmarks.landmark[16].x*image_width), int(results.pose_landmarks.landmark[16].y*image_height))
      leftShoulder = (int(results.pose_landmarks.landmark[11].x*image_width), int(results.pose_landmarks.landmark[11].y*image_height))
      rightShoulder = (int(results.pose_landmarks.landmark[12].x*image_width), int(results.pose_landmarks.landmark[12].y*image_height))

    
      if distanceCalculate(rightShoulder,rightWrist)<200 or distanceCalculate(leftShoulder,leftWrist)<200:
          pushUpStart = 1
      elif pushUpStart and (distanceCalculate(rightShoulder,rightWrist)>maxDistance or distanceCalculate(leftShoulder,leftWrist)>maxDistance ):
          pushUpCount += 1
          pushUpStart = 0

      print(pushUpCount)
      print(distanceCalculate(rightShoulder,rightWrist))
      print(distanceCalculate(leftShoulder,leftWrist))

      font = cv2.FONT_HERSHEY_SIMPLEX
      org = (50, 100)
      fontScale = 2
      color = (0, 0, 0)
      color1 = (255, 255, 255)
      thickness = 8
      thickness1 = 2
      annotated_image = cv2.putText(image, "Pompki counter:  " + str(pushUpCount), org, font, fontScale, color, thickness)
      annotated_image = cv2.putText(image, "Pompki counter:  " + str(pushUpCount), org, font, fontScale, color1, thickness1)
      mp_drawing.draw_landmarks(
          annotated_image,
          results.pose_landmarks,
          mp_pose.POSE_CONNECTIONS)


      result.write(annotated_image)


  cap.release()
```

```python
path1 = "/content/drive/MyDrive/BOKS/pompki.mp4"
res1 = (1920,1080)
name1 = "pompki"
maxDistance1 = 700
path2 = "/content/drive/MyDrive/BOKS/pompki1.mp4"
res2 = (2560,1440)
name2 = "pompki1"
maxDistance2 = 600
pushUpCounter(path1,res1,name1,maxDistance1)
pushUpCounter(path2,res2,name2,maxDistance2)
```
